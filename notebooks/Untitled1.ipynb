```python
output = []
for info in glob.glob('B:\\Projects\\maude\\*\\*.json'):
    data = json.load(open(info))
    found = False
    for line in data['description'].split('\n'):
        if re.search('(writ?ten|by\s|writer)',line,flags=re.I) and not re.search('(Filmed|Direct)',line, flags=re.I):
            found = True
            output.append((info.split('\\')[3],data['fulltitle'],line,data['webpage_url']))
            break
    if not found:
        output.append((info.split('\\')[3],data['fulltitle'],'No writer',data['webpage_url']))
        
```

```python
df = pd.DataFrame(output)
df.columns = ['team','sketch_title','writer','link']
df.pipe(remove_writer)
df.to_csv('maude.csv',encoding ='utf-8')
# df.to_clipboard(index=False,sep=';')
```

```python
def remove_writer(df):
    df['writer_clean'] = df['writer'].apply(lambda x: re.sub('(writ?ten|by?:\s|writer)','',x,flags=re.I).strip())
    df['writer_clean'] = df['writer_clean'].apply(lambda x: re.sub('by[\s:]','',x,flags=re.I).strip())
    return df
```
